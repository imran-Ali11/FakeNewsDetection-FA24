```python
def preprocess_urdu_text(text):
    text = re.sub(r'[^\u0600-\u06FF\s]', '', text)  # Keep only Urdu characters
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
    return text
```
